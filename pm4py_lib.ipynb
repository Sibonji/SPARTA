```python
import pandas
import pm4py
from sberpm import DataHolder

file_path = "data_table.csv"

event_log = pandas.read_csv(file_path, encoding="cp1251")
event_log = pm4py.format_dataframe(event_log, case_id='case', activity_key='stage', start_timestamp_key='start_time', timestamp_key='end_time', timest_format='%d.%m.%Y %H:%M')
start_activities = pm4py.get_start_activities(event_log)
end_activities = pm4py.get_end_activities(event_log)
print("Start activities: {}\nEnd activities: {}".format(start_activities, end_activities))
```
